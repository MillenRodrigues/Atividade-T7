Com base no dataset específico **pontuacaoJogadas**, apresente os dados somente quando ouve *rush* ou *pass* na jogada. Garanta que exista também o nome ou abreviatura do time que está atacando (variável *posteam*), além dos nomes dos jogadores que estão fazendo passe,  correndo ou recebendo a bola (variáveis *passer*, *rusher* e *receiver*)

Utilizando o subconjunto de dados **acoesJogadas** e **pontuacaoJogadas**, crie uma análise que retorne qual foi o jogador que conquistou mais jardas no terceiro quarto.

```r
# cruza as bases e filtra
pontuacoesTerceiroQuarto <-
  pontuacaoJogadas %>%  # base com nomes dos jogadores
  inner_join(acoesJogadas, by='play_id') %>% # junta base que possui jardas e quarter
  filter(qtr == 3) %>% # filtra jogadas do terceiro quarter
  filter(yards_gained != 0) %>% # apenas as jogadas em que houve ganho ou perda de jardas
  select(yards_gained, name) %>% # só nos interessa o nome do jogador e jardas
  group_by(name) %>% # agrupa por jogador
  summarise(yards=sum(yards_gained)) %>% # soma as jardas
  arrange(desc(yards)) # ordena decrescente para facilitar a leitura

# exibe os resultados
pontuacoesTerceiroQuarto
```

#### Desafio de geração de gráfico  

Crie um gráfico de linhas, mostrando a pontuação de cada time em cada *quarter*. O resultado deve ter duas linhas, uma para cada time, e cada linha será composta pela pontuação de cada um dos *quarters* sendo uma cor para cada time. O eixo X terá os *quarters* e o eixo y terá a pontuação.

```r
# construção da base
df <-
  pontuacaoJogadas %>%  # base com pontuações
  inner_join(acoesJogadas, by='play_id') %>% # junta base que indica o quarter
  filter(!is.na(posteam)) %>% # filtra deixando apenas jogadas que indicam time atacante
  group_by(posteam, qtr) %>% # agrupa por time (atacante) e quarter
  summarise(score=max(posteam_score),  # seleciona o score máximo para cada time em cada quarter
    .groups='keep')  # evita warning sobre agrupamento com parâmetro padrão


# plota o gráfico
ggplot(data=df, aes(x=qtr, y=score, group=posteam)) + # define parametros do gráfico
  geom_line(aes(color=posteam))+ # adiciona as linhas e define a cor por time
  geom_point(aes(color=posteam)) + # adiciona pontos sobre a linha da cor de cada time
  theme(text = element_text(size = 20)) + # aumenta o texto que é pequeno no colab
  labs(title = "Pontuação por Quarter", # adiciona um título ao gráfico
    x = "Quarter", y = "Pontuação", color="Time") # renomeia os eixos e legenda de cores

```

Crie um gráfico de barras empilhada (colunas verticais), utilizando somente as jogadas que tiveram entre 10 e 20 jardas conquistas.
O empilhamento das barras será feito pela quantidade de jardas conquistadas (entre 10 e 20). Mantenha as barras verticais segmentadas por quarter do jogo, e por fim, crie a faceta baseada nos times.


```r
# construção da base
df <-
  pontuacaoJogadas %>%  # base com nomes de times
  inner_join(acoesJogadas, by='play_id') %>% # junta base que indica o quarter e jardas
  filter(10 <= yards_gained & yards_gained <= 20) %>% # filtra deixando apenas jogadas com jardas entre 10 e 20
  select(posteam, yards_gained, qtr) # seleciona apenas as colunas relevantes para o grafico



# plota o gráfico
ggplot(data=df, # define os dados
  aes(x=qtr, fill=yards_gained, group=yards_gained)) + # define x para quarters e y para jardas
  geom_bar(position=position_stack(reverse = TRUE)) + # define formato de barras empilhadas
  facet_grid("posteam") +  # cria faceta por time
  theme(text = element_text(size = 20)) + # aumenta o texto que é pequeno no colab
  labs(title = "Jogada em jardas por Quarter", # adiciona um título ao gráfico
        x = "Quarter", y = "Jogadas", fill="Jardas") + # renomeia os eixos
  scale_fill_gradient(low="blue", high="red") # define a escala de cores do azul pro vermelho
```
